```python
import matplotlib.pyplot as plt
import nltk
import pandas as pd
import re
import seaborn as sns

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split

# Load data
fake_df = pd.read_csv("Fake.csv")
real_df = pd.read_csv("True.csv")

# Add class label
fake_df["label"] = 1
real_df["label"] = 0

# Sampling
fake_sample = fake_df.sample(n=3000, random_state=42)
real_sample = real_df.sample(n=7000, random_state=42)

# Combine data
data = pd.concat([fake_sample, real_sample]).sample(frac=1, random_state=42).reset_index(drop=True)
X = data["text"]
y = data["label"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.3, random_state=42)

# Download resources
nltk.download("wordnet")
nltk.download('punkt_tab')
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("omw-1.4")

# Lemmatization function
wnl = WordNetLemmatizer()
stop_words = set(stopwords.words("english"))

def lemmatize_text(text):
    text = re.sub(r"[^a-zA-Z]", " ", text).lower()
    tokens = nltk.word_tokenize(text)
    lemmatized = [wnl.lemmatize(token) for token in tokens if token not in stop_words]
    return " ".join(lemmatized)

# Apply lemmatization
X_train_lemmatized = X_train.apply(lemmatize_text)
X_test_lemmatized = X_test.apply(lemmatize_text)

# Vocabulary sizes
vocab_size_before = len(CountVectorizer().fit(X_train).vocabulary_)
vocab_size_after = len(CountVectorizer().fit(X_train_lemmatized).vocabulary_)

print("Vocabulary before preprocessing:", vocab_size_before)
print("Vocabulary after preprocessing:", vocab_size_after)


# TF-IDF representation
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train_lemmatized)
X_test_tfidf = tfidf_vectorizer.transform(X_test_lemmatized)

# Binary representation
count_vectorizer = CountVectorizer(binary=True)
X_train_binary = count_vectorizer.fit_transform(X_train_lemmatized)
X_test_binary = count_vectorizer.transform(X_test_lemmatized)

# Parameter settings
param_grid = [
    {"C": 0.1, "max_iter": 100},
    {"C": 1.0, "max_iter": 100},
    {"C": 5.0, "max_iter": 100},
    {"C": 10.0, "max_iter": 100},
    {"C": 15.0, "max_iter": 100},
]



def run_experiments(X_train, X_test, y_train, y_test, representation_name):
    results = []

    print(f"Representation: {representation_name}")
    for i, params in enumerate(param_grid, start=1):
        print(f"Experiment {i}, Parameters: {params}")
        model = LogisticRegression(**params, random_state=42)
        model.fit(X_train, y_train)

        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)

        # Metrics
        metrics = {
            "experiment": i,
            "params": params,
            "representation": representation_name,

            # Training
            "train_accuracy": accuracy_score(y_train, y_pred_train),
            "train_precision": precision_score(y_train, y_pred_train),
            "train_recall": recall_score(y_train, y_pred_train),
            "train_f1": f1_score(y_train, y_pred_train),

            # Test
            "test_accuracy": accuracy_score(y_test, y_pred_test),
            "test_precision": precision_score(y_test, y_pred_test),
            "test_recall": recall_score(y_test, y_pred_test),
            "test_f1": f1_score(y_test, y_pred_test),
        }
        results.append(metrics)

        print("Training:")
        print(classification_report(y_train, y_pred_train, digits=4))

        print("Test:")
        print(classification_report(y_test, y_pred_test, digits=4))

        # Confusion matrix
        cm = confusion_matrix(y_test, y_pred_test)
        plt.figure(figsize=(5, 4))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        plt.title(f"Confusion matrix, {representation_name}, Experiment {i}")
        plt.xlabel("Prediction")
        plt.ylabel("Real")
        plt.show()

    return pd.DataFrame(results)

# Run experiments for both representations
results_tfidf = run_experiments(X_train_tfidf, X_test_tfidf, y_train, y_test, "TF-IDF")
results_binary = run_experiments(X_train_binary, X_test_binary, y_train, y_test, "Binary")

all_results = pd.concat([results_tfidf, results_binary], ignore_index=True)
all_results.sort_values(by="test_f1", ascending=False)
```
